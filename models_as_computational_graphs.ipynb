```python
# A logistic regression forward pass
import torch.nn.functional as F
import torch

y = torch.tensor([1.0])
x1 = torch.tensor([1.1])
w1 = torch.tensor([2.2])
b = torch.tensor([0.0])
# Forward pass
z = w1 * x1 + b
a = torch.sigmoid(z)
print(f"Output: {a.item()}")  # Print the output value
loss = F.binary_cross_entropy(a, y)
print(f"Loss: {loss.item()}")  # Print the loss value
```
